```python
lhs_scaled = pyDOE2.lhs(2, criterion='centermaximin', samples=36)

levels = np.array([6, 6])
design = pyDOE2.fullfact(levels)
levels[levels == 1] = 2
ff_scaled = design / (levels - 1)

pts_lhs = list(zip(*lhs_scaled))
pts_ff = list(zip(*ff_scaled))

fig = go.Figure()

fig.add_trace(go.Scatter(x=pts_lhs[0], y=pts_lhs[1], mode='markers', showlegend=False))
fig.add_trace(go.Scatter(x=pts_ff[0], y=pts_ff[1], mode='markers', showlegend=False))

fig.show()
```
